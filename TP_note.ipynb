```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage.util import random_noise
```

```python
def medianFilter(img,k):
  w,h,c = img.shape
  size = k // 2
    #0 processus de remplissage
  _img = np.zeros((w+2*size,h+2*size,c), dtype=np.float)
  _img[size:size+w,size:size+h] = img.copy().astype(np.float)
  dst = _img.copy()
#Processus de filtrage
  for x in range(w):
    for y in range(h):
      for z in range(c):
        dst[x+size,y+size,z] = np.median(_img[x:x+k,y:y+k,z])

  dst = dst[size:size+w,size:size+h].astype(np.uint8)

  return dst
```

```python
#Lecture d'image
img = cv2.imread('Image_tpNote.jpg')

#Filtre mdian
#Deuxime argument: taille du filtre
img = medianFilter(img,15)
#Enregistrer l'image
cv2.imwrite('Imageresult.jpg', img)
#Affichage de l'image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
